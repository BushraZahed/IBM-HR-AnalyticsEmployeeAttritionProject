### <b>Let's Get Started</b>
We use Pandas and Numpy Libraries for data manipulation & calculation and Matplotlib and Seaborn libraries for data visualization. 
## 1. Importing Libraies

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline 
import seaborn as sns
```

## 2. Loading Data into Pandas Dataframe

```python
df = pd.read_csv("WA_Fn-UseC_-HR-Employee-Attrition-clean.csv")
df.head()
```

## 3. Which education levels have higher attrition rate?

```python
df.Education.value_counts()  # 1. below college, 2. College, 3. Bachelors, 4. Masters, 5. PhD
```

```python
edu=df.groupby(['Attrition', 'Education']).size().unstack() #0 means attrition yes , 1 means attrition no
edu
```

```python
# calculating the percentage for each education level
data_percent = edu.apply(lambda x: round(100*x/x.sum(),2), axis=1)
data_percent
```

```python
# Using the colorblind palette from the Seaborn library for visualization
colors = sns.color_palette('colorblind')
# create a stacked bar plot
ax = data_percent.plot(kind='bar', stacked=False, figsize=(10,10))

# setting the title and axis labels
ax.set_title('Attrition Rate by Education Level')
ax.set_xlabel('Education Level')
ax.set_ylabel('Attrition Rate (%)')

# adding legend
handles, labels = ax.get_legend_handles_labels()
ax.legend(handles[::-1], labels[::-1], title='Attrition')

# showing the plot
plt.show()
```

Based on the given attrition rates, it appears that employees with education level 3 have the highest attrition rate (41.77%), followed by those with level 4 (24.47%), level 2 (18.57%), level 1 (13.08%), and level 5 (2.11%). Based on the data provided, employees with higher education levels such as Bachelor's and Master's degrees have a higher attrition rate compared to those with lower education levels. This could be due to higher expectations, more opportunities, mismatched skills, or a desire for work-life balance. To address employee retention, companies should provide opportunities for career growth and development, offer competitive compensation and benefits, promote a positive work culture, and provide training and support to improve skills and knowledge. As for why employees with PhDs may have a lower attrition rate, it's possible that they have more specialized skills and knowledge that are in higher demand in the job market, and may therefore have more opportunities available to them within the company. Previously, from univariate analysis we have seen that the nature of the company is "Researh-Oriented" and the R & D dept has more employees than any other departments. As such the company may value employees with PhDs and offer more attractive compensation and benefits packages to retain them. Finally, employees with PhDs may have different career goals or priorities that make them less likely to leave, such as a focus on research within the company.
